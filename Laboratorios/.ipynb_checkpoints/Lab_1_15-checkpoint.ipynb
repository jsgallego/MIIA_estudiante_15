* *Milla extra*: escribe una función que sea capaz de hacer lo anterior para un archivo y úsala en un ciclo para automatizar la lectura de los diferentes archivos.

```python
def read_str (t):
    
    with open(t,mode="r", encoding="utf-8") as f:
        f = f.read()
        f = f.splitlines()        
    return f


def read_float (t):
    
    with open(t,mode="r", encoding="utf-8") as f:
        f = f.read()
        f = f.splitlines()
        f = [float(i) for i in f]
    return f


txt = ["edad", "escolaridad", "estado_civil", "estrato", "genero", "promedio", "region"]


for i in range(len(txt)):
    if txt[i] == "edad":
        exec(txt[i]+'{} = read_float ("./Archivos/"+ txt[i] +".txt")'.format(i))
    elif txt[i] == "promedio":
        exec(txt[i]+'{} = read_float ("./Archivos/"+ txt[i] +".txt")'.format(i))
    elif txt[i] == "estrato":
        exec(txt[i]+'{} = read_float ("./Archivos/"+ txt[i] +".txt")'.format(i))
    else:
        exec(txt[i]+'{} = read_str ("./Archivos/"+ txt[i] +".txt")'.format(i))


datos2 = {"edad" : list(edad0),
         "escolaridad" : list(escolaridad1),
         "estado_civil" : list(estado_civil2),
         "estrato" : list(estrato3),
         "genero" : list(genero4),
         "promedio" : list(promedio5),
         "region" : list(region6)}

```

**Misión 4**

Desarrollar un algoritmo que permita conocer la población que sería becada según los criterios propuestos por el *primer grupo* mencionado arriba.

```python
# b =  número de becas a asignar

# arroja dos diccionarios, uno en donde cada llave es un individuo becado, con su respectivo índice dentro de los 
# datos originales, para que el comité encargado sepa exactamente a quiénes se les asignaron las becas y cuáles son sus 
# características

# y otro donde cada llave es una variable, y su valor es la lista de todos los becados 
# y los individuos son los índices de cada lista.

# se decide mantener el 2% de cada subpoblación original, para que el número de becas asignadas en cada ciclo 
# se mantenga constante

def becados_g1(b):
    
    b_promedio5 = list(promedio5)
    
    b1 = []
    b2 = []
    b3 = []
    b4 = []
    b5 = []
    n1 = len(list(filter(lambda x: x == 1, datos2["estrato"])))
    n2 = len(list(filter(lambda x: x == 2, datos2["estrato"])))
    n3 = len(list(filter(lambda x: x == 3, datos2["estrato"])))
    n4 = len(list(filter(lambda x: x == 4, datos2["estrato"])))
    n5 = len(list(filter(lambda x: x == 5, datos2["estrato"])))
    ciclo = 1
    
    while (len(b1)+len(b2)+len(b3)+len(b4)+len(b5)) < b:
        
        # estrato 1
        for i in range(len(edad0)):
            if b1.count(i) == 0:
                if datos2["estrato"][i] == 1.0:
                    promedio_1 =[]
                    
                    for j in range(len(edad0)):
                        if datos2["estrato"][j] == 1.0:
                            promedio_1.append(b_promedio5[j])
                    
                    if b_promedio5[i] == max(promedio_1):
                        b1.append(i)
                        b_promedio5[i] = 0
                    
                    if (len(b1)+len(b2)+len(b3)+len(b4)+len(b5)) == b:   ## afinar
                        break
                    else:
                        continue
                
                if len(b1) >= int(n1*(0.02)*ciclo):   ## afinar
                    break
                else:
                    continue
            
            
            
        # estrato 2    
        for i in range(len(edad0)):
            if b2.count(i) == 0:
                if datos2["estrato"][i] == 2.0:
                    promedio_2 =[]
                    
                    for j in range(len(edad0)):
                        if datos2["estrato"][j] == 2.0:
                            promedio_2.append(b_promedio5[j])
                    
                    if b_promedio5[i] == max(promedio_2):
                        b2.append(i)
                        b_promedio5[i] = 0
                        
                    if (len(b1)+len(b2)+len(b3)+len(b4)+len(b5)) == b:   ## afinar
                        break
                    else:
                        continue
                
                if len(b2) >= int(n2*(0.02)*ciclo):   ## afinar
                    break
                else:
                    continue
            
            
        # estrato 3    
        for i in range(len(edad0)):
            if b3.count(i) == 0:
                if datos2["estrato"][i] == 3.0:
                    promedio_3 =[]
                    
                    for j in range(len(edad0)):
                        if datos2["estrato"][j] == 3.0:
                            promedio_3.append(b_promedio5[j])
                    
                    if b_promedio5[i] == max(promedio_3):
                        b3.append(i)
                        b_promedio5[i] = 0
                    
                    if (len(b1)+len(b2)+len(b3)+len(b4)+len(b5)) == b:   ## afinar
                        break
                    else:
                        continue
                
                if len(b3) >= int(n3*(0.02)*ciclo):   ## afinar
                    break
                else:
                    continue
       
        # estrato 4    
        for i in range(len(edad0)):
            if b4.count(i) == 0:
                if datos2["estrato"][i] == 4.0:
                    promedio_4 =[]
                    
                    for j in range(len(edad0)):
                        if datos2["estrato"][j] == 4.0:
                            promedio_4.append(b_promedio5[j])
                    
                    if b_promedio5[i] == max(promedio_4):
                        b4.append(i)
                        b_promedio5[i] = 0
                    
                    if (len(b1)+len(b2)+len(b3)+len(b4)+len(b5)) == b:   ## afinar
                        break
                    else:
                        continue
                
                if len(b4) == int(n4*(0.02)*ciclo):   ## afinar
                    break
                else:
                    continue
        
        
        # estrato 5    
        for i in range(len(edad0)): 
            
            if b5.count(i) == 0:
                if datos2["estrato"][i] == 5.0:
                    promedio_5 =[]
                    
                    for j in range(len(edad0)):
                        if datos2["estrato"][j] == 5.0:
                            promedio_5.append(b_promedio5[j])
                    
                    if b_promedio5[i] == max(promedio_5):
                        b5.append(i)
                        b_promedio5[i] = 0
                        
                    if (len(b1)+len(b2)+len(b3)+len(b4)+len(b5)) == b:   ## afinar
                        break
                    else:
                        continue
                    
                if len(b5) >= int(n5*(0.02)*ciclo):   ## afinar
                    break
                else:
                    continue
        
        ciclo += 1     
        #fin ciclo while
        
    i_becas = b1 + b2 + b3 + b4 + b5
        
    becados = {}
    
    for i in i_becas:
        becados[i] =[datos2["edad"][i],
                    datos2["escolaridad"][i],
                    datos2["estado_civil"][i],
                    datos2["estrato"][i],
                    datos2["genero"][i],
                    datos2["promedio"][i],
                    datos2["region"][i]]
                
    edad = []
    escolaridad = []
    estado_civil = []
    estrato = []
    genero = []
    promedio = []
    region = []
    
    
    for i in i_becas:
        edad.append(datos2["edad"][i])
        escolaridad.append(datos2["escolaridad"][i])
        estado_civil.append(datos2["estado_civil"][i])
        estrato.append(datos2["estrato"][i])
        genero.append(datos2["genero"][i])
        promedio.append(datos2["promedio"][i])
        region.append(datos2["region"][i])
    
    
    
    lista_becados = {"edad" : list(edad),
                     "escolaridad" : list(escolaridad),
                     "estado_civil" : list(estado_civil),
                     "estrato" : list(estrato),
                     "genero" : list(genero),
                     "promedio" : list(promedio),
                     "region" : list(region)}

    
    return becados, lista_becados
```

```python
# b =  número de becas a asignar

# arroja dos diccionarios, uno en donde cada llave es un individuo becado, con su respectivo índice dentro de los 
# datos originales, para que el comité encargado sepa exactamente a quiénes se les asignaron las becas y cuáles son sus 
# características

# y otro donde cada llave es una variable, y su valor es la lista de todos los becados 
# y los individuos son los índices de cada lista.

# Teniendo en cuenta que la condición es que las becas deben estar igualmente distribuidas entre las diferentes regiones 
# y generos, y no proporcionalmente o equitativamente distribuidas, se configura el algoritmo para que cada subgrupo tenga
# el mismo número de becas. Si hay subgrupos con poblaciones menores al número de becas asignadas a dicha población, entonces
# esas becas sobrantes quedan sin asignar.

def becados_g2(b):
    c_genero = ["femenino", "masculino", "no binario", "otro"]
    c_region = ["Amazonia", "Andina", "Caribe", "Orinoquia", "Pacifica"]
    
    c = []
    for i in c_genero:
            for j in c_region:
                c.append ([i,j])
                
     
    b_promedio = list(promedio5)
    
    
    b_ =  [[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[]]
    
    promedio_ =  [[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[]]
    
        
    numero_becas_grupo = b / len(c)
    
    for j in range(len(c)):
        for i in range(len(edad0)):
            if datos2["genero"][i] == c[j][0]:
                if datos2["region"][i] == c[j][1]:
                    b_[j].append(i)
                    promedio_[j].append(b_promedio[i])
    
    
    
    
    becas = [[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[],[]]
    ciclo = 1
    l = len(becas[0])+len(becas[1])+len(becas[2])+len(becas[3])+len(becas[4])+len(becas[5])+len(becas[6])+len(becas[7])+len(becas[8])+len(becas[9])+len(becas[10])+len(becas[11])+len(becas[12])+len(becas[13])+len(becas[14])+len(becas[15])+len(becas[16])+len(becas[17])+len(becas[18])+len(becas[19])
    
    while (l/20) < int(numero_becas_grupo):
        
        for j in range(len(promedio_)):
            l = len(becas[0])+len(becas[1])+len(becas[2])+len(becas[3])+len(becas[4])+len(becas[5])+len(becas[6])+len(becas[7])+len(becas[8])+len(becas[9])+len(becas[10])+len(becas[11])+len(becas[12])+len(becas[13])+len(becas[14])+len(becas[15])+len(becas[16])+len(becas[17])+len(becas[18])+len(becas[19])
            for i in range(len(promedio_[j])):
    
                if promedio_[j][i] == max(promedio_[j]):
                    becas[j].append(b_[j][i])
                    promedio_[j][i] = 0
    
                    if len(becas[j]) == int(numero_becas_grupo):   ## afinar
                        break
                    else:
                        continue
                
                if len(becas[j]) == int(numero_becas_grupo):   ## afinar
                    break
                else:
                    continue
            if l/20 >= int(numero_becas_grupo):   ## afinar
                break
            else:
                continue
    
            
            ciclo += 1     
            #fin ciclo while
    
    index_becas = []
    for n in range(len(becas)):
        index_becas += becas[n]
    
    index_becas = list(set(index_becas))
    
    becados = {}
        
    for i in index_becas:
        becados[i] =[datos2["edad"][i],
                    datos2["escolaridad"][i],
                    datos2["estado_civil"][i],
                    datos2["estrato"][i],
                    datos2["genero"][i],
                    datos2["promedio"][i],
                    datos2["region"][i]]
                
    edad = []
    escolaridad = []
    estado_civil = []
    estrato = []
    genero = []
    promedio = []
    region = []
    
    
    for i in index_becas:
        edad.append(datos2["edad"][i])
        escolaridad.append(datos2["escolaridad"][i])
        estado_civil.append(datos2["estado_civil"][i])
        estrato.append(datos2["estrato"][i])
        genero.append(datos2["genero"][i])
        promedio.append(datos2["promedio"][i])
        region.append(datos2["region"][i])
    
    
    
    lista_becados = {"edad" : list(edad),
                     "escolaridad" : list(escolaridad),
                     "estado_civil" : list(estado_civil),
                     "estrato" : list(estrato),
                     "genero" : list(genero),
                     "promedio" : list(promedio),
                     "region" : list(region)}
    
    return becados, lista_becados
```

Reconociendo la complejidad de este problema (que no permite reducirlo a una dicotomía) y viendo el potencial del análisis de datos para enriquecer la discusión y los insumos para tomar decisiones, el comité ha solicitado una herramienta analítica que les permita . Sus requerimientos son:

**Misión 7**

Desarrollar, a partir de lo construido hasta el momento, un algoritmo que permita cumplir con los requerimientos del comité.

```python
# Función la cual tiene los siguientes parámetros:
# n_becas = número de becas a asignar
# criterio = critero único de asignación (genero, estrato, region)
# proporcion = lista con la proporcion de cada categoría dentro del criterio. Cada proporción es un número entre 0 y 1.
#              la suma de todas las proporciones debe ser igual a 1
#              Si es género [femenino, masculino, no binario, otro], si es estrato [1, 2, 3, 4, 5]
#              Si es region [Amazonia, Andina, Caribe, Orinoquia, Pacifica]
# min_edad = el mínimo de edad para que sea aplicable a la beca
# max_edad = el máximo de edad para que sea aplicable a la beca

def becados(n_becas, criterio, proporcion, min_edad=0, max_edad=200):
    
    import math 
    if criterio != "genero" and criterio !="estrato" and criterio !="region":
        return "Por favor seleccione un criterio válido, entre genero, estrato o region"
        
    if math.fsum(proporcion) != 1:
        return "Por favor utilice una lista con proporciones válidas, que sumen 1" 
    
    b_persona = []
    b_edad = []
    b_escolaridad = []
    b_estado_civil = []
    b_estrato = []
    b_genero = []
    b_promedio = []
    b_region = []
    
    for i in range(len(promedio5)):
        if datos2["edad"][i] >= min_edad and datos2["edad"][i] <= max_edad:
                b_persona.append(i)
                b_edad.append(datos2["edad"][i])
                b_escolaridad.append(datos2["escolaridad"][i])
                b_estado_civil.append(datos2["estado_civil"][i])
                b_estrato.append(datos2["estrato"][i])
                b_genero.append(datos2["genero"][i])
                b_promedio.append(datos2["promedio"][i])
                b_region.append(datos2["region"][i])
    
    ### genero
    if criterio == "genero":
        if len(proporcion) != 4:
            return "Por favor utilice una lista con cuatro proporciones, una para cada genero"
        else:
            b_=[[],[],[],[]]
            promedio_=[[],[],[],[]]
            c = ["femenino", "masculino", "no binario", "otro"]
            
            for j in range(len(c)):
                for i in range(len(b_edad)):
                    if b_genero[i] == c[j]:
                        b_[j].append(b_persona[i])
                        promedio_[j].append(b_promedio[i])
            
            becas = [[],[],[],[]]
            ciclo = 1
                  
            for j in range(len(promedio_)):
                while len(becas[j]) < int(n_becas*proporcion[j]) or ciclo >= 1000000:
                    m = max(promedio_[j])
                    if type(m) == float:
                        indice = promedio_[j].index(m)
                        becas[j].append(b_[j][indice])
                        promedio_[j][indice]=-1      
                    ciclo += 1
                        
                    if ciclo == 1000000:
                        break
    
    ### region
    if criterio == "region":
        if len(proporcion) != 5:
            return "Por favor utilice una lista con cinco proporciones, una para cada region"
        else:
            b_=[[],[],[],[],[]]
            promedio_=[[],[],[],[],[]]
            c = ["Amazonia", "Andina", "Caribe", "Orinoquia", "Pacifica"]
            
            for j in range(len(c)):
                for i in range(len(b_edad)):
                    if b_region[i] == c[j]:
                        b_[j].append(b_persona[i])
                        promedio_[j].append(b_promedio[i])
            
            becas = [[],[],[],[],[]]
            ciclo = 1
                  
            for j in range(len(promedio_)):
                while len(becas[j]) < int(n_becas*proporcion[j]) or ciclo >= 1000000:
                    m = max(promedio_[j])
                    if type(m) == float:
                        indice = promedio_[j].index(m)
                        becas[j].append(b_[j][indice])
                        promedio_[j][indice]=-1      
                    ciclo += 1
                        
                    if ciclo == 1000000:
                        break
                
    ## estrato
    if criterio == "estrato":
        if len(proporcion) != 5:
            return "Por favor utilice una lista con cinco proporciones, una para cada estrato"
        else:
            b_=[[],[],[],[],[]]
            promedio_=[[],[],[],[],[]]
            c = [1, 2, 3, 4, 5]
            
            for j in range(len(c)):
                for i in range(len(b_edad)):
                    if b_estrato[i] == c[j]:
                        b_[j].append(b_persona[i])
                        promedio_[j].append(b_promedio[i])
            
            becas = [[],[],[],[],[]]
            ciclo = 1
                  
            for j in range(len(promedio_)):
                while len(becas[j]) < int(n_becas*proporcion[j]) or ciclo >= 1000000:
                    m = max(promedio_[j])
                    if type(m) == float:
                        indice = promedio_[j].index(m)
                        becas[j].append(b_[j][indice])
                        promedio_[j][indice]=-1      
                    ciclo += 1
                        
                    if ciclo == 1000000:
                        break  

    index_becas = []
    for n in range(len(becas)):
        index_becas += becas[n]
    
    
    becados = {}
        
    for i in index_becas:
        becados[i] =[datos2["edad"][i],
                    datos2["escolaridad"][i],
                    datos2["estado_civil"][i],
                    datos2["estrato"][i],
                    datos2["genero"][i],
                    datos2["promedio"][i],
                    datos2["region"][i]]
                
    edad = []
    escolaridad = []
    estado_civil = []
    estrato = []
    genero = []
    promedio = []
    region = []
    
    
    for i in index_becas:
        edad.append(datos2["edad"][i])
        escolaridad.append(datos2["escolaridad"][i])
        estado_civil.append(datos2["estado_civil"][i])
        estrato.append(datos2["estrato"][i])
        genero.append(datos2["genero"][i])
        promedio.append(datos2["promedio"][i])
        region.append(datos2["region"][i])
    
    
    
    lista_becados = {"edad" : list(edad),
                     "escolaridad" : list(escolaridad),
                     "estado_civil" : list(estado_civil),
                     "estrato" : list(estrato),
                     "genero" : list(genero),
                     "promedio" : list(promedio),
                     "region" : list(region)}


    return becados, lista_becados
    
```
